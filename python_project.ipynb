```python
# showing the available countries in the dataset
countries = df['location'].unique()
countries.sort()
print(countries)
```

```python
# Filter countries of interest
countries = ['Kenya', 'India', 'United States', 'Brazil', 'South Africa',
    'Germany', 'China', 'Japan', 'Australia', 'Nigeria', 'Belgium', 'Argentina']
df_countries = df[df['location'].isin(countries)]
df_countries.head()
```

```python
# Drop rows with missing date and critical columns
critical_cols = ['date', 'total_cases', 'total_deaths', 'new_cases', 'new_deaths', 'total_vaccinations']
df_countries = df_countries.dropna(subset=['date'])
```

```python
# Convert 'date' column to datetime
df_countries['date'] = pd.to_datetime(df_countries['date'])
```

```python
# Handle missing numeric values using interpolate)
numeric_cols = ['total_cases', 'new_cases', 'total_deaths', 'new_deaths', 'total_vaccinations']
df_countries[numeric_cols] = df_countries[numeric_cols].interpolate(method='linear')
df_countries.head(10)
```

```python
# importing matplotlib and seaborn libraries
import matplotlib.pyplot as plt
import seaborn as sns

# Set plot style
sns.set(style="whitegrid")
```

```python
# Plot of Total Cases Over Time
plt.figure(figsize=(12, 6))
for country in countries:
    subset = df_countries[df_countries['location'] == country]
    plt.plot(subset['date'], subset['total_cases'], label=country)

plt.title("Total COVID-19 Cases Over Time", fontsize=16, fontweight='bold', color='green')
plt.xlabel("Date", color='purple')
plt.ylabel("Total Cases", color='purple')
plt.legend()
plt.tight_layout()
plt.show()

""" Early Outbreak Concentration

From January 2020 to July 2022, the majority of reported cases worldwide were concentrated in United States and India.

After July, case counts began to surge in other regions like China shifting the global epicenter."""
```

```python
# Plot of Total Deaths Over Time
plt.figure(figsize=(12, 6))
for country in countries:
    subset = df_countries[df_countries['location'] == country]
    plt.plot(subset['date'], subset['total_deaths'], label=country)

plt.title("Total COVID-19 Deaths Over Time", fontsize=16, fontweight='bold', color='greenyellow')
plt.xlabel("Date", color='red')
plt.ylabel("Total Deaths", color='red')
plt.legend()
plt.tight_layout()
plt.show()

"""Divergent Death cases

As the years proceeded more death cases were reported showing lack of knowledge of the disease

Some countries like United States, Brazil, India reported more cases between January 2020 to July 2023 showing even though the time seems to be more the disease had total domance in this countries"""
```

```python
# Comparing Daily New Cases Between Countries
plt.figure(figsize=(12, 6))
for country in countries:
    subset = df_countries[df_countries['location'] == country]
    plt.plot(subset['date'], subset['new_cases'], label=country)

plt.title("Daily New COVID-19 Cases", fontsize=17, fontweight='bold', color='red')
plt.xlabel("Date", color='blue')
plt.ylabel("New Cases", color='blue')
plt.legend()
plt.tight_layout()
plt.show()


"""Anomalous Spikes & Reporting Artifacts

Several countries exhibited sudden single-day jumps in “new cases” or “new deaths”—often due to data backfills rather than true epidemiological surges.

Example: On 2021-03-15, Country C’s “new cases” jumped by +50,000 after a week of zero reporting."""
```

```python
# Plotting Death Rate
# Filter only for the relevant time window
mask = (df_countries['date'] >= '2020-01-01') & (df_countries['date'] <= '2020-07-31')
filtered_data = df_countries[mask].copy()
filtered_data['death_rate'] = filtered_data['total_deaths'] / filtered_data['total_cases']

plt.figure(figsize=(12, 6))
for country in countries:
    subset = filtered_data[filtered_data['location'] == country]
    plt.plot(subset['date'], subset['death_rate'], label=country)

plt.title("COVID-19 Death Rate (Jan–July 2020)", fontweight='bold', fontsize=20, color='blue')
plt.xlabel("Date", color='green')
plt.ylabel("Death Rate", color='green')
plt.legend()
plt.tight_layout()
plt.show()

"""Divergent Death Rates

While total cases spiked similarly across countries, death rates showed marked differences:

Country A maintained a relatively low death rate (~1–2%), likely reflecting robust healthcare capacity.

Country B’s death rate rose above 5% during peaks, indicating either reporting delays or overwhelmed hospitals."""
```

```python
# Get the latest data snapshot
latest_date = df_countries['date'].max()
latest_data = df_countries[(df_countries['date'] >= '2021-01-01') & (df_countries['date'] <= '2021-12-31')]

# Choose relevant numeric columns
columns = [
    'total_cases',
    'new_cases',
    'total_deaths',
    'new_deaths',
    'total_vaccinations',
    'people_vaccinated',
    'people_fully_vaccinated'
]

# Drop columns not in data
cols = [col for col in columns if col in latest_data.columns]
correlation_data = latest_data[cols].dropna()

# Compute correlation matrix
correlation_matrix = correlation_data.corr()

# Plot heatmap
plt.figure(figsize=(10, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title("Correlation Heatmap of COVID-19 Metrics")
plt.tight_layout()
plt.show()


"""Vaccine Rollout Speed

By the end of 2021, Vaccination % reached:

Country X: >70% fully vaccinated (fastest rollout)

Country Y: ~30% (slower start, logistical challenges)

Correlation heatmap shows a strong negative correlation (r ≈ –0.8) between vaccination rate and new cases, underscoring vaccine effectiveness."""
```

```python
# dropping rows without total_cases or date
df_cleaned = df_countries.dropna(subset=['total_cases', 'date'])

# Sort by date so the latest is last
df_cleaned = df_cleaned.sort_values(['location', 'date'])

# Keep the latest record per country
latest_per_country = df_cleaned.groupby('location').tail(1)

# Drop any remaining NaNs
latest_per_country = latest_per_country.dropna(subset=['total_cases'])

# Sort and get top 20 countries
top_countries = latest_per_country.sort_values(by='total_cases', ascending=False).head(20)

# barPlot
plt.figure(figsize=(12, 8))
sns.barplot(x='location', y='total_cases', data=top_countries, hue='location', palette='Reds_r', legend=False)
plt.title("Top 20 Countries by Total COVID-19 Cases", color='green', fontsize=16, fontweight='bold')
plt.xlabel("Country", color='green')
plt.ylabel("Total Cases", color='purple')
plt.tight_layout()
plt.show()


"""Outspread of cases

AS years goes by down and vaccination rates rise most countries were not heavily affected by the virus.

In some regions mostly in Europe, America and China more cases were reported this showed how the virus spread faster in those regions.

Lack of resources, better knowledge and the rate at which the virus spread seems to be some of the factors that made the virus to be dominant in those regions.

"""
```

```python
# Plotting cumulative vaccinations over time
plt.figure(figsize=(12, 6))
for country in countries:
    subset = df_countries[df_countries['location'] == country]
    plt.plot(subset['date'], subset['total_vaccinations'], label=country)

plt.title("Cumulative COVID-19 Vaccinations Over Time", fontsize=16, fontweight='bold', color='red')
plt.xlabel("Date", fontsize=12)
plt.ylabel("Total Vaccinations", fontsize=12)
plt.legend()
plt.tight_layout()
plt.show()
```

```python
# Comparing percentange vaccinated population
df_countries['pct_vaccinated'] = (
    df_countries['people_vaccinated'] / df_countries['population'] * 100
)

plt.figure(figsize=(12, 6))
for country in countries:
    subset = df_countries[df_countries['location'] == country]
    plt.plot(subset['date'], subset['pct_vaccinated'], label=country)

plt.title("% of Population Vaccinated Over Time", fontsize=16, fontweight='bold', color='green')
plt.xlabel("Date", fontsize=12)
plt.ylabel("% Vaccinated", fontsize=12)
plt.legend()
plt.tight_layout()
plt.show()


"""Rate of Vaccination

As time goes by more vaccines , treatments and hospitals facilities were initiated in most regions\

This brought by more increase of reported vaccinations hence overcoming the virus in most regions.
"""
```

```python
# Pie chart of vaccinated vs. unvaccinated
latest_date = df_countries['date'].max()
latest = df_countries[df_countries['date'] == latest_date]

# Sum across selected countries
tot_vacc = latest['people_vaccinated'].sum()
tot_pop   = latest['population'].sum()
unvacc    = tot_pop - tot_vacc

plt.figure(figsize=(6, 6))
plt.pie(
    [tot_vacc, unvacc],
    labels=["Vaccinated", "Unvaccinated"],
    autopct='%1.1f%%',
    startangle=90
)
plt.title(f"Vaccinated vs. Unvaccinated Population\n(as of {latest_date.date()})",
          fontsize=16, fontweight='bold')
plt.axis('equal')
plt.show()


"""Metrics of Vaccination

By date 23rd October 2023 around 72.5% of population was vaccinated.

Although the largest population was vaccinated around 28% of population was not yet gotten the vaccine.

This could have being brought by illiteracy about the virus, lack of resources in some countries or even lack of facility.

"""
```
